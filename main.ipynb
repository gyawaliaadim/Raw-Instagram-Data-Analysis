```python
import json

with open("finaldata.txt", encoding='utf-8') as f:
    data = f.read()
```

```python
def validateTotalNumber(value):
    if "K" in value.upper():
        return int(float(value.upper().replace("K",""))*1000)
    if "M" in value.upper():
        return int(float(value.upper().replace("M",""))*1000000)
    if "B" in value.upper():
        return int(float(value.upper().replace("B",""))*1000000000)
    else:
        return int(value)
```

```python
def parseChunk(chunk):
    splittedChunks=chunk.split("\n")
    username=splittedChunks[0]
    posts=validateTotalNumber(splittedChunks[1].replace(" posts","").replace(" post","").replace(",",""))
    followers=validateTotalNumber(splittedChunks[2].replace(" followers","").replace(" follower","").replace(",",""))
    # followers=splittedChunks[2]
    following=validateTotalNumber(splittedChunks[3].replace(" following","").replace(",",""))
    name=splittedChunks[4]
    if len(splittedChunks)>=6:
        type_of_page=splittedChunks[5]
        bio="\n".join(splittedChunks[6:])
        return {"username":username,"name": name, "posts": posts, "followers": followers, "following":following,"type":type_of_page,"bio": bio}
   
    bio ="\n".join(splittedChunks[5:])    
    return {"username":username,"name": name, "posts": posts, "followers": followers, "following":following,"type":"","bio": bio}
    
```

```python

def findMaxOfCheckKey(chunks,checkKey):
    maxOfKey=0
    chunkWithMaxOfKey=None
    for chunk in chunks:
        if maxOfKey<chunk[checkKey]:
            maxOfKey=chunk[checkKey]
            chunkWithMaxOfKey=chunk
    return chunkWithMaxOfKey
     
```

```python
chunks = data.split("\n\n")
chunks = [chunk.strip() for chunk in chunks if chunk.strip()!=""]
parsedData=[]
for chunk in chunks:
    parsedData.append(parseChunk(chunk))
    # print(chunk)
s=json.dumps(parsedData,indent=4)
# print(s)
# for account in parsedData:
#     dictAccount=dict(account)
print(f"Most Posts is of {findMaxOfCheckKey(parsedData,"posts")}\n")
print(f"Most Followers is of {findMaxOfCheckKey(parsedData,"followers")}\n")
print(f"Most Following is of {findMaxOfCheckKey(parsedData,"following")}\n")
catogories=[]
for chunk in parsedData:
    if chunk["type"] not in catogories and chunk["type"].strip()!="":
        catogories.append(chunk["type"])
#         pass
        # print(chunk["type"])
print(f"There are in total {len(catogories)} catogories.\n")

print(f"They are {", ".join(catogories)}")
#     # print(account)
#     print(dict(account)["followers"])
#     print(dict(account)["posts"])
```
